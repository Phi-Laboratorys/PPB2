```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import openpyxl
import xlsxwriter
from pandas_ods_reader import read_ods
```

```python
data1 = "Messung3_3_SiMono_dunkel.ods"
data2 = "Messung3_3_SiMono_130.ods"
data3 = "Messung3_3_SiMono_180.ods"
data4 = "Messung3_3_SiMono_230.ods"
data5 = "Messung3_3_SiMulti_dunkel.ods"
data6 = "Messung3_3_SiMulti_130.ods"
data7 = "Messung3_3_SiMulti_180.ods"
data8 = "Messung3_3_SiMulti_230.ods"
data9 = "Messung3_3_CIS_dunkel.ods"
data10 = "Messung3_3_CIS_130.ods"
data11 = "Messung3_3_CIS_180.ods"
data12 = "Messung3_3_CIS_230.ods"
```
